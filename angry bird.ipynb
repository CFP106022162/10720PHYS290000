```python
import numpy as np
import matplotlib.pyplot as plt
import math

v = 50
theta = 45
g = -9.8
vx = v*math.sin(theta)
vy = v*math.cos(theta)

t = np.arange( 0 , -2*vy/g , 0.01 )
xx = t * vx ;
xy = vy * t + 0.5 * g * t * t ;
plt.plot(xx,xy)

x_pig = np.random.uniform(0,20)
y_pig = np.random.uniform(0,30)
plt.plot(x_pig,y_pig,"o")
```
